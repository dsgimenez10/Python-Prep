```python
# Crear una lista
mi_lista = [1, 2, 3, 4, 5]
print(f"Lista original: {mi_lista}")

# Acceder a elementos de una lista
primer_elemento = mi_lista[0]
ultimo_elemento = mi_lista[-1]
print(f"Primer elemento: {primer_elemento}")
print(f"Último elemento: {ultimo_elemento}")

# Modificar elementos de una lista
mi_lista[1] = 10
print(f"Lista después de modificar el segundo elemento: {mi_lista}")

# Añadir elementos a una lista
mi_lista.append(6)
print(f"Lista después de añadir el elemento 6: {mi_lista}")

# Longitud de una lista
longitud = len(mi_lista)
print(f"Longitud de la lista: {longitud}")

# Iterar sobre una lista
print("Elementos de la lista:")
for elemento in mi_lista:
    print(elemento)

# Eliminar elementos de una lista
mi_lista.remove(3)
print(f"Lista después de eliminar el primer elemento con valor 3: {mi_lista}")

# Operaciones con listas
otra_lista = [7, 8, 9]
concatenacion = mi_lista + otra_lista
repetir = mi_lista * 2
print(f"Concatenación de listas: {concatenacion}")
print(f"Repetir la lista dos veces: {repetir}")

# Listas anidadas
lista_anidada = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(f"Lista anidada: {lista_anidada}")

# Métodos útiles de listas
mi_lista.sort()
print(f"Lista ordenada de forma ascendente: {mi_lista}")

mi_lista.reverse()
print(f"Lista invertida: {mi_lista}")

# Agregar un elemento al final de la lista
mi_lista.append(7)
print(f"Lista después de añadir el elemento 7 al final: {mi_lista}")

# Insertar un elemento en una posición específica
mi_lista.insert(2, 99)
print(f"Lista después de insertar el número 99 en la posición 2: {mi_lista}")

# Extender la lista con otra lista
mi_lista.extend([8, 9, 10])
print(f"Lista después de extenderla con [8, 9, 10]: {mi_lista}")

# Eliminar el último elemento de la lista y devolverlo
ultimo_elemento = mi_lista.pop()
print(f"Lista después de usar pop para eliminar y devolver el último elemento: {mi_lista}")
print(f"Elemento eliminado con pop: {ultimo_elemento}")

# Eliminar un elemento en una posición específica
mi_lista.pop(2)
print(f"Lista después de usar pop para eliminar el elemento en la posición 2: {mi_lista}")

# Eliminar el primer elemento con un valor específico
mi_lista.remove(9)
print(f"Lista después de usar remove para eliminar el primer elemento con valor 9: {mi_lista}")

# Copiar una lista
copia_lista = mi_lista.copy()
print(f"Copia de la lista original: {copia_lista}")

# Limpiar todos los elementos de la lista
mi_lista.clear()
print(f"Lista después de limpiar todos los elementos: {mi_lista}")
```

```python
# Crear una tupla
mi_tupla = (1, 2, 3, 4, 5)
print(f"Tupla original: {mi_tupla}")

# Acceder a elementos de una tupla
primer_elemento = mi_tupla[0]
ultimo_elemento = mi_tupla[-1]
print(f"Primer elemento: {primer_elemento}")
print(f"Último elemento: {ultimo_elemento}")

# Longitud de una tupla
longitud = len(mi_tupla)
print(f"Longitud de la tupla: {longitud}")

# Iterar sobre una tupla
print("Elementos de la tupla:")
for elemento in mi_tupla:
    print(elemento)

# Concatenación de tuplas
otra_tupla = (6, 7, 8)
concatenacion = mi_tupla + otra_tupla
print(f"Concatenación de tuplas: {concatenacion}")

# Repetir una tupla
repetir = mi_tupla * 2
print(f"Repetir la tupla dos veces: {repetir}")

# Encontrar el índice de un elemento
indice_3 = mi_tupla.index(3)
print(f"Índice del elemento 3: {indice_3}")

# Contar la cantidad de veces que aparece un elemento
conteo_4 = mi_tupla.count(4)
print(f"Conteo del elemento 4: {conteo_4}")

# Desempaquetado de tuplas
tupla_desempaquetada = (10, 20, 30)
a, b, c = tupla_desempaquetada
print(f"Desempaquetado: a={a}, b={b}, c={c}")

```

```python
# Crear un diccionario
mi_diccionario = {"nombre": "Juan", "edad": 25, "ciudad": "Ejemplo"}

# Acceder a valores mediante claves
valor_nombre = mi_diccionario["nombre"]
valor_edad = mi_diccionario["edad"]
valor_ciudad = mi_diccionario["ciudad"]

# Modificar un valor
mi_diccionario["edad"] = 26

# Añadir un nuevo par clave-valor
mi_diccionario["profesion"] = "Desarrollador"

# Eliminar un par clave-valor
del mi_diccionario["ciudad"]

# Obtener todas las claves
claves = mi_diccionario.keys()

# Obtener todos los valores
valores = mi_diccionario.values()

# Obtener pares clave-valor como tuplas
items = mi_diccionario.items()

# Comprobar si una clave está en el diccionario
existe_clave = "nombre" in mi_diccionario

# Iterar sobre las claves
print("Claves:")
for clave in mi_diccionario:
    print(clave)

# Iterar sobre los valores
print("Valores:")
for valor in mi_diccionario.values():
    print(valor)

# Iterar sobre pares clave-valor
print("Pares clave-valor:")
for clave, valor in mi_diccionario.items():
    print(f"Clave: {clave}, Valor: {valor}")

# Diccionarios anidados
diccionario_anidado = {"persona": {"nombre": "Ana", "edad": 30}, "producto": {"nombre": "Laptop", "precio": 1200}}

# Acceder a un valor en un diccionario anidado
valor_nombre_persona = diccionario_anidado["persona"]["nombre"]

# Imprimir resultados
print(f"Valor de 'nombre': {valor_nombre}")
print(f"Valor de 'edad' (modificado): {mi_diccionario['edad']}")
print(f"Diccionario después de añadir 'profesion': {mi_diccionario}")
print(f"Claves: {claves}")
print(f"Valores: {valores}")
print(f"Pares clave-valor: {items}")
print(f"¿Existe la clave 'nombre'?: {existe_clave}")

# Imprimir diccionario anidado
print(f"\nDiccionario anidado:\n{diccionario_anidado}")
print(f"Valor de 'nombre' en 'persona': {valor_nombre_persona}")
```
