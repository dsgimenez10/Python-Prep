```python
# Operaciones con strings

# Concatenación de strings
cadena1 = "Hola"
cadena2 = " Mundo"
concatenacion = cadena1 + cadena2

# Repetición de strings
repeticion = cadena1 * 3  # Repetir la cadena1 tres veces

# Acceso a caracteres por índice
primer_caracter = cadena1[0]  # Obtener el primer carácter

# Longitud de una cadena
longitud = len(cadena1)

# Métodos de formato
mayusculas = cadena1.upper()
minusculas = cadena1.lower()
capitalizado = cadena1.capitalize()

# Reemplazo de caracteres
reemplazo = cadena1.replace("o", "0")

# Separación de strings
frase = "Hola Mundo, cómo estás"
palabras = frase.split()  # Por defecto, se separa por espacios en blanco

# Unión de strings
lista_de_palabras = ["Hola", "Mundo", "Python"]
union = " ".join(lista_de_palabras)  # Unir elementos de la lista con un espacio en blanco

# Eliminación de espacios en blanco
cadena_espacios = "   Esto es una cadena con espacios   "
sin_espacios_inicio_fin = cadena_espacios.strip()

# Formato con placeholders
nombre = "Juan"
edad = 25
mensaje = "Hola, me llamo {} y tengo {} años".format(nombre, edad)

# F-strings (formato literal de cadena)
mensaje_fstring = f"Hola, me llamo {nombre} y tengo {edad} años"

# Imprimir resultados
print(f"Concatenación: {concatenacion}")
print(f"Repetición: {repeticion}")
print(f"Primer carácter: {primer_caracter}")
print(f"Longitud: {longitud}")
print(f"Mayúsculas: {mayusculas}")
print(f"Minúsculas: {minusculas}")
print(f"Capitalizado: {capitalizado}")
print(f"Reemplazo: {reemplazo}")
print(f"Palabras separadas: {palabras}")
print(f"Unión de palabras: {union}")
print(f"Sin espacios al inicio y al final: '{sin_espacios_inicio_fin}'")
print(f"Mensaje formateado: {mensaje}")
print(f"Mensaje con f-string: {mensaje_fstring}")

# Operaciones con strings (continuación)

# Búsqueda de subcadenas
frase = "Hola Mundo, cómo estás"
busqueda = "Mundo" in frase  # Devuelve True si "Mundo" está en la frase

# Búsqueda y obtención de índices
indice_mundo = frase.find("Mundo")  # Devuelve el índice de la primera aparición de "Mundo"
ultimo_indice_o = frase.rfind("o")  # Devuelve el índice de la última aparición de "o"

# Verificación de contenido
es_digito = frase.isdigit()  # Devuelve True si todos los caracteres son dígitos
es_alpha = frase.isalpha()  # Devuelve True si todos los caracteres son alfabéticos
es_espacio = frase.isspace()  # Devuelve True si todos los caracteres son espacios en blanco

# Slicing (corte de cadenas)
subcadena = frase[5:10]  # Obtiene los caracteres desde el índice 5 hasta el 9

# Conteo de apariciones
conteo_o = frase.count("o")  # Cuenta cuántas veces aparece "o" en la frase

# Capitalización de palabras
frase_titulo = frase.title()  # Convierte la primera letra de cada palabra a mayúscula

# Reversión de una cadena
cadena_reversa = frase[::-1]

# Formateo de números en una cadena
numero = 42
cadena_formateada = "El número es {:03d}".format(numero)  # Rellena con ceros a la izquierda

# F-strings para formateo de cadenas
nombre = "Alice"
edad = 30
mensaje_fstring = f"Hola, me llamo {nombre.upper()} y tengo {edad + 5} años"

# Imprimir resultados
print(f"Búsqueda de subcadena: {busqueda}")
print(f"Índice de 'Mundo': {indice_mundo}")
print(f"Último índice de 'o': {ultimo_indice_o}")
print(f"Es un número: {es_digito}")
print(f"Es alfabético: {es_alpha}")
print(f"Es espacio en blanco: {es_espacio}")
print(f"Slicing: {subcadena}")
print(f"Conteo de 'o': {conteo_o}")
print(f"Frase en título: {frase_titulo}")
print(f"Cadena reversa: {cadena_reversa}")
print(f"Cadena formateada: {cadena_formateada}")
print(f"Mensaje con f-string: {mensaje_fstring}")
```

```python
# Operaciones lógicas

# Operadores de comparación
numero1 = 10
numero2 = 5

igualdad = numero1 == numero2
desigualdad = numero1 != numero2
mayor_que = numero1 > numero2
menor_que = numero1 < numero2
mayor_o_igual_que = numero1 >= numero2
menor_o_igual_que = numero1 <= numero2

# Operadores lógicos
and_logico = (numero1 > 5) and (numero2 < 8)  # True si ambas condiciones son True
or_logico = (numero1 > 5) or (numero2 > 8)    # True si al menos una condición es True
not_logico = not (numero1 > 5)                # True si la condición es False

# Operaciones con booleanos
booleano1 = True
booleano2 = False

and_booleano = booleano1 and booleano2  # False si al menos uno es False
or_booleano = booleano1 or booleano2    # True si al menos uno es True
not_booleano1 = not booleano1           # Invierte el valor lógico

# Imprimir resultados
print(f"Igualdad: {igualdad}")
print(f"Desigualdad: {desigualdad}")
print(f"Mayor que: {mayor_que}")
print(f"Menor que: {menor_que}")
print(f"Mayor o igual que: {mayor_o_igual_que}")
print(f"Menor o igual que: {menor_o_igual_que}")
print(f"And lógico: {and_logico}")
print(f"Or lógico: {or_logico}")
print(f"Not lógico: {not_logico}")
print(f"And booleano: {and_booleano}")
print(f"Or booleano: {or_booleano}")
print(f"Not booleano1: {not_booleano1}")
```

```python
# Casting de tipo de datos

# De cadena a entero
cadena_numero = "123"
entero = int(cadena_numero)

# De entero a cadena
numero = 456
cadena_numero = str(numero)

# De entero a flotante
entero = 789
flotante = float(entero)

# De flotante a entero
flotante = 123.45
entero = int(flotante)

# Imprimir resultados
print(f"De cadena a entero: {entero}")
print(f"De entero a cadena: {cadena_numero}")
print(f"De entero a flotante: {flotante}")
print(f"De flotante a entero: {entero}")

# Conversión de cadena binaria a entero
binario = "1010"
entero_desde_binario = int(binario, 2)

# Imprimir resultado
print(f"De binario a entero: {entero_desde_binario}")
```

```python
# Representación de números binarios

# Números binarios se prefijan con '0b'
binario1 = 0b1010  # Representa el número decimal 10 en binario
binario2 = 0b1100  # Representa el número decimal 12 en binario

# Imprimir resultados
print(f"Binario 1: {binario1}")
print(f"Binario 2: {binario2}")
```

```python
# Operaciones de bits

# Operador OR a nivel de bits
resultado_or = binario1 | binario2

# Operador AND a nivel de bits
resultado_and = binario1 & binario2

# Operador XOR a nivel de bits
resultado_xor = binario1 ^ binario2

# Desplazamiento a la izquierda (multiplicar por 2)
resultado_desplazamiento_izquierda = binario1 << 1

# Desplazamiento a la derecha (dividir por 2)
resultado_desplazamiento_derecha = binario2 >> 1

# Imprimir resultados
print(f"OR a nivel de bits: {resultado_or}")
print(f"AND a nivel de bits: {resultado_and}")
print(f"XOR a nivel de bits: {resultado_xor}")
print(f"Desplazamiento a la izquierda: {resultado_desplazamiento_izquierda}")
print(f"Desplazamiento a la derecha: {resultado_desplazamiento_derecha}")

# Uso de funciones incorporadas para trabajar con binarios
cadena_binaria = "1010"
entero_desde_binario = int(cadena_binaria, 2)

# Imprimir resultado
print(f"De binario a entero: {entero_desde_binario}")

```
