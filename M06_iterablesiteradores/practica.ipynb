```python
# Listas
lista_numeros = [1, 2, 3, 4, 5]
lista_letras = ['a', 'b', 'c', 'd', 'e']

# Tuplas
tupla_colores = ('rojo', 'verde', 'azul')

# Conjuntos
conjunto_frutas = {'manzana', 'banana', 'naranja'}

# Diccionario
diccionario = {'a': 1, 'b': 2, 'c': 3}

# Cadena de texto
cadena_saludo = "Hola, mundo!"

# Iterar sobre listas
for numero in lista_numeros:
    print(numero)

for letra in lista_letras:
    print(letra)

# Iterar sobre tuplas
for color in tupla_colores:
    print(color)

# Iterar sobre conjuntos
for fruta in conjunto_frutas:
    print(fruta)

# Iterar sobre diccionario
for clave, valor in diccionario.items():
    print(clave, valor)

# Iterar sobre cadena de texto
for caracter in cadena_saludo:
    print(caracter)

# Utilizando enumerate
for indice, valor in enumerate(lista_letras):
    print(f"Índice: {indice}, Valor: {valor}")

# Utilizando zip
numeros = [1, 2, 3]
letras = ['a', 'b', 'c']

for numero, letra in zip(numeros, letras):
    print(numero, letra)

# Iteración condicional
pares = [x for x in lista_numeros if x % 2 == 0]
impares = [x for x in lista_numeros if x % 2 != 0]

print("Pares:", pares)
print("Impares:", impares)
```
